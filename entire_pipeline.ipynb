```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.imp
```

```python
df=pd.read_csv('/Users/atul2501/Desktop/train.csv')
```

```python
df.head(20)
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head(5)
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=40)
```

```python
X_train.isnull().sum()
```

```python
X_test
```

```python
y_train.sample(5)
```

```python
#imputer 
trf1=ColumnTransformer([
    ('imputer_age',SimpleImputer(),[2]),
    ('impuer_embarked',SimpleImputer(strategy='most_frequent'),[6])
    
],remainder='passthrough')
```

```python
#OneHotIncoding
trf2=ColumnTransformer([
    ('Ohe_sex_emberked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
    
],remainder='passthrough')
```

```python
#scaled for better performace
trf3=ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
#Feature selection 
trf4=SelectKBest(score_func=chi2,k=8)
```

```python
trf5=DecisionTreeClassifier()
```

```python
#Create Pipeline
pipe=Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
    
])
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.named_steps['trf1'].transformers_[1][1].statistics_
```

```python
#Predict
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#Cross Validation using pipeline
from sklearn.model_selection import cross_val_score
cross_val_score(pipe,X_train,y_train,cv=100,scoring='accuracy').mean()
```

```python
#GridSerch using Pipeline
params={
    'trf5__max_depth':[1,2,3,4,5,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(pipe,params,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```
